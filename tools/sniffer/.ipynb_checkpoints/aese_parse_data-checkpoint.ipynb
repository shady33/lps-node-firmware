```python
def split_data(element):
    split_data1 = element.split(' ')[0:4]
    parsed_data = []
    for i in split_data1:
        s=binascii.unhexlify(i[2:-1])
        parsed_data.append(struct.unpack('>I',s)[0])
    split_data2 = list(map(float,element.split(' ')[4:]))
    return parsed_data + split_data2
    
values = list(map(split_data,data[::2]))
z = pd.DataFrame(values)
```

```python
data = []
with open('collect_some_data.txt') as f:
    data = f.readline()
```

```python
i = -9999999
data_of_interest = []
second_data_of_interest = []
poll_tx = 0
poll_rx = 0
answer_tx = 0
answer_rx = 0
final_tx = 0
final_rx = 0
df = pd.DataFrame(columns=['poll_tx','poll_rx','answer_tx','answer_rx','final_tx','final_rx'])
for line in data.split('\\n'):
    if line[0] == 'I':
                    
        new_row = {'poll_tx':poll_tx,'poll_rx':poll_rx,'answer_tx':answer_tx,'answer_rx':answer_rx,'final_tx':final_tx,'final_rx':final_rx}
        df = df.append(new_row,ignore_index=True)
        
        d1 = []
        for val in line[1:].split(':'):
            d1.append(struct.unpack('>I',binascii.unhexlify(val))[0])
        second_data_of_interest.append(d1)
        i = 0
    else:
        if i == 0:
            poll_tx = struct.unpack('>I',binascii.unhexlify(line.split(' ')[0][2:]))[0]
            poll_rx = struct.unpack('>I',binascii.unhexlify(line.split(' ')[1][2:]))[0]
        if i == 1:
            answer_tx = struct.unpack('>I',binascii.unhexlify(line.split(' ')[0][2:]))[0]
            answer_rx = struct.unpack('>I',binascii.unhexlify(line.split(' ')[1][2:]))[0]
        if i == 2:
            final_tx = struct.unpack('>I',binascii.unhexlify(line.split(' ')[0][2:]))[0]
            final_rx =struct.unpack('>I',binascii.unhexlify(line.split(' ')[1][2:]))[0]
        i = i + 1
df = df.iloc[1:]
```

```python
tround1 = df['answer_rx'] - df['poll_tx']
treply1 = df['answer_tx'] - df['poll_rx']
tround2 = df['final_rx'] - df['answer_tx']
treply2 = df['final_tx'] - df['answer_rx']
```

```python
C = 299792458.0
tsfreq = 499.2e6 * 128

tprop_ctn = (((tround1*tround2) - (treply1*treply2)) / (tround1 + tround2 + treply1 + treply2))
tprop = tprop_ctn/tsfreq
distance = C * tprop
px.line(distance)
```
